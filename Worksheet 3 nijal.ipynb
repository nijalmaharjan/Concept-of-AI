```python
# Problem 1 - Sorting:
import pandas as pd
titanic_df = pd.read_csv('Titanic-Dataset.csv')

# Check info of the DataFrame
print("Dataset Info:")
print(titanic_df.info())
print("\nFirst 5 rows:")
print(titanic_df.head()) # prints the first 5 rows.

#1. Create a DataFrame called fare that contains only the Fare column
fare = titanic_df[['Fare']] # creates dataframe Fare
print("Fare DataFrame (first 5 rows):")
print(fare.head())

#2. Create a DataFrame called class_age that contains Pclass and Age columns
class_age = titanic_df[['Pclass', 'Age']] #creating the dataframe pclass and age
print("\nClass and Age DataFrame (first 5 rows):")
print(class_age.head()) # printing the first 5 rows by .head

#3. Create a DataFrame called survived_gender that contains Survived and Sex columns
survived_gender = titanic_df[['Survived', 'Sex']]
print("\nSurvived and Gender DataFrame (first 5 rows):")
print(survived_gender.head()) # printing teh first 5 rows
```

```python
# Problem 2 - Subsetting (selecting specific rows)

# 1. Select all passengers who paid more than 100 as ticket fare
fare_gt_100 = titanic_df[titanic_df['Fare'] > 100] # if fare is greter than 100 vanyeko

# Print how many passengers paid more than 100
print(f"Passengers with fare > 100: {len(fare_gt_100)}") # len funciton counts

# Show first 5 rows of this filtered data
print("First 5 rows:")
print(fare_gt_100.head()) #head slows 5 rowss


# 2. Select all passengers who are in first class (Pclass = 1)
first_class = titanic_df[titanic_df['Pclass'] == 1] # checking passengers who are in 1st class

# Print number of first class passengers
print(f"\nFirst class passengers: {len(first_class)}")

# Show first 5 rows of first class passengers
print("First 5 rows:")
print(first_class.head())


# 3. Select all female passengers under age 18
female_under_18 = titanic_df[(titanic_df['Age'] < 18) & (titanic_df['Sex'] == 'female')] # & oprator so duetai conditon true huna paryu

# Print number of female passengers under 18
print(f"\nFemale passengers under 18: {len(female_under_18)}") # again len counts

# Show first 5 rows of filtered results
print("First 5 rows:")
print(female_under_18.head()) # prints 5 rows
```

```python
#Subsetting Rows by Categorical variables:
# 1. Filter for passengers embarked at 'C' or 'S'
embarked_c_or_s = titanic_df[titanic_df['Embarked'].isin(['C', 'S'])] #isin checks if there is C,S xa ki xiana
print(f"Passengers embarked at C or S: {len(embarked_c_or_s)}") # len will check katiota xa
print("First 5 rows:")
print(embarked_c_or_s.head())

# 2. Filter for first or second class passengers
first_second_class = titanic_df[titanic_df['Pclass'].isin([1, 2])] # isin, xa ki nai check
print(f"\nFirst or second class passengers: {len(first_second_class)}")
print("First 5 rows:")
print(first_second_class.head())
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise - 1
# Which passenger had the highest fare paid relative to their age?
# Create a copy of the dataframe to avoid SettingWithCopyWarning
titanic_eda = titanic_df.copy() # make a copy of a dataset so it wont affect the original data

# Handle missing values in Age column by filling with median age
median_age = titanic_eda['Age'].median() #finds median of the age (bichako)
titanic_eda['Age'] = titanic_eda['Age'].fillna(median_age) # fillna() removes the Nan value and make the value x

# Add fare_per_year column
titanic_eda['fare_per_year'] = titanic_eda['Fare'] / titanic_eda['Age']

# Subset rows where fare_per_year > 5
high_fare_age = titanic_eda[titanic_eda['fare_per_year'] > 5] #checkingg..........

# Sort by descending fare_per_year
high_fare_age_srt = high_fare_age.sort_values('fare_per_year', ascending=False) #decending order (high fare to low)

# Select only Name and fare_per_year columns
result = high_fare_age_srt[['Name', 'fare_per_year']]

print("Passengers with highest fare per year (first 10):")
print(result.head(10)) # 10 ota rows

print(f"\nThe passenger with highest fare relative to age is: {result.iloc[0]['Name']}") # iloc vanyeko selecting by number and loc -> by label
print(f"With fare per year of: {result.iloc[0]['fare_per_year']:.2f}") #.2f means decimal paxhi 2 ta digit
```

```python
# Which adult male passenger paid the highest fare relative to their class?
# Add fare_per_class column
titanic_eda['fare_per_class'] = titanic_eda['Fare'] / titanic_eda['Pclass']

# Subset for adult male passengers
adult_males = titanic_eda[(titanic_eda['Sex'] == 'male') & (titanic_eda['Age'] >= 18)]

# Sort by descending fare_per_class
adult_males_srt = adult_males.sort_values('fare_per_class', ascending=False) # high fare to low

# Select required columns
result_males = adult_males_srt[['Name', 'Age', 'fare_per_class']]

print("Adult male passengers with highest fare per class (first 10):")
print(result_males.head(10))

print(f"\nThe adult male passenger with highest fare relative to class is: {result_males.iloc[0]['Name']}")
print(f"Age: {result_males.iloc[0]['Age']}, Fare per class: {result_males.iloc[0]['fare_per_class']:.2f}") # selecting row 0 of Age and fareperclass
```

```python
# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise
# What percent of the total fare revenue came from each passenger class?
# Calculate total fare
total_fare = titanic_df['Fare'].sum() # adds up all the fare
print(f"Total fare revenue: {total_fare:.2f}") #decimal paxhi 2 ta num matra

# Calculate fare by class
fare_by_class = titanic_df.groupby('Pclass')['Fare'].sum() # pclass ko sum like 1 5000, 2 1000 , 3 200
print("\nFare revenue by class:")
print(fare_by_class)

# Calculate percentage
percentage_by_class = (fare_by_class / total_fare) * 100 # calculating in percentage
print("\nPercentage of total fare revenue by class:")
for pclass, percentage in percentage_by_class.items():
    print(f"Class {pclass}: {percentage:.2f}%")
```

```python
# What percent of the total number of passengers belonged to each age group?
# Create age_group column
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 65:
        return 'Adult'
    else:
        return 'Senior'

titanic_df['age_group'] = titanic_df['Age'].apply(categorize_age) #calling mathi ko function

# Calculate total number of passengers
total_passengers = len(titanic_df) # len counts the row
print(f"Total passengers: {total_passengers}")

# Count passengers in each age group
age_group_counts = titanic_df['age_group'].value_counts() # counts how many passenger are in each group
print("\nPassengers by age group:")
print(age_group_counts)

# Calculate percentage
age_group_percentage = (age_group_counts / total_passengers) * 100
print("\nPercentage of passengers by age group:")
for group, percentage in age_group_percentage.items():
    print(f"{group}: {percentage:.2f}%")

```
